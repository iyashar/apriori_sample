```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules,fpgrowth,fpmax
```

```python
data = pd.read_csv('supermarket.csv')
```

```python
data.head()
```

```python
# فرض می کنیم خرید هر مشتری در هر روز یک سبد خرید است. لذا ستون جدیدی با عنوان سبد به داده ها اضافه می کنیم
data['Sabad']=data['Customer Id']+data['Date']
```

```python
data.info()
```

```python
data.describe()
```

```python
data2 = data.groupby(['Sabad','Product']).size()
datar = pd.DataFrame(data2).reset_index()
datar.columns = ['Sabad','Product','Count']
datar.head()

```

```python
Tdata = (datar.groupby(['Sabad', 'Product'])['Count']
          .sum().unstack().reset_index().fillna(0)
          .set_index('Sabad'))
Tdata.head()
```

```python
def hot_encode(x):
    if(x<= 0):
        return 0
    if(x>= 1):
        return 1
 
# Encoding the datasets
Tdata_encoded = Tdata.applymap(hot_encode)
Tdata_encoded.head()
```

```python
# Building the model
frq_items = apriori(Tdata_encoded, min_support = 0.01, use_colnames = True).sort_values('support', ascending = False)
print(frq_items)
```

```python
# Collecting the inferred rules in a dataframe
rules = association_rules(frq_items, metric ="lift", min_threshold = 0.001)
rules = rules.sort_values(['confidence'], ascending =[False])
print(rules)
```
